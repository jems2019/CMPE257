```
url = "https://www.wunderground.com/history/monthly/us/ca/napa/KSTS/date/{}"
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')


weather_df = pd.DataFrame(columns=["Year", "Average Temperature", "Average Rainfall"])

```

```
import time
```

```
for year in range(1990,1991):
  temp = 0
  rain = 0
  wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
  for month in range(1,12):
    dateStr = str(year) +"-"+ str(month)
    urlNew = url.format(dateStr)
    wd.implicitly_wait(20)
    wd.get(urlNew)
    time.sleep(15)
    sample_soup = BeautifulSoup(wd.page_source, 'html.parser')
    summary_table = sample_soup.find(class_="summary-table")
    rows = summary_table.find_all("td")
    temp += float(rows[10].text)     
    rain += float(rows[23].text)
  weather_df = weather_df.append({'Year' : int(year) , 'Average Temperature' : temp/12, 'Average Rainfall' : rain/12 } , ignore_index=True)
      

    
    
```
