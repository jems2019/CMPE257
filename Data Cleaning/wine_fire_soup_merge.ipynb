```python
import pandas as pd
import numpy as np
```

```python
wine_data = pd.read_csv('../data/wine_fire_weather_data.csv')
soup_data = pd.read_csv('../data/data.csv')
```

```python
soup_data.head()
soup_data.drop_duplicates(inplace=True)
```

```python
wine_data['ratings'] = 0
wine_data['reviews'] = 0
for index, row in wine_data.iterrows():
    try:
        wine_data.loc[index, 'ratings'] = float(soup_data.loc[soup_data.loc[:,'Winery'] == row['winery'],'Ratings'])
        wine_data.loc[index, 'reviews'] = float(soup_data.loc[soup_data.loc[:,'Winery'] == row['winery'],'Number of Reviews'])
    except:
        wine_data.loc[index, 'ratings'] = float('NaN')
        wine_data.loc[index, 'reviews'] = float('NaN')
wine_data.head()

```

```python
print wine_data.notna().sum()
wine_data.shape
```

```python
wine_data.dropna(inplace=True)
```

```python
wine_data['norm_ratings'] = abs((wine_data.ratings-wine_data.ratings.min()) / wine_data.ratings.max())
wine_data['norm_reviews'] = abs((wine_data.reviews-wine_data.reviews.min()) / wine_data.reviews.max())
```

```python
wine_data.shape
```

```python
wine_data['review_words'] = 0

for index, row in wine_data.iterrows():
   wine_data.loc[index, 'review_words'] = row['description'].count(' ')

wine_data.head()
```

```python
wine_data.to_csv('../data/wine_fire_soup.csv')
```
